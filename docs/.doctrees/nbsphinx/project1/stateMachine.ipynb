````C++
/// ISEL-LEFA-OEF3, 14/09/2025
//Codigo exemplo 
// maquina de estados - sistema multitarefa
//Manfred Niehus, manfred.niehus@isel.pt

#define LED1 2
#define LED2 3
float f1=1.0;  //f1 freqencia LED1
float f2=2.0;  //f2 freqencia LED2
int dutycycle=50;  //duty cycle (%)

void setup(){
  Serial.begin(115200);
  pinMode(LED1,OUTPUT);
  pinMode(LED2,OUTPUT);
}

void loop(){
  autoBlinkLed1(LED1,f1,dutycycle);
  autoBlinkLed2(LED2,f2,dutycycle);
}

// state machine 1
void autoBlinkLed1(int pin,float f, int duty){
  // pisca um LED no pin com frequencia f
  // Input parameters: 
  //      int pin - numero pin do arduino - entre [2-13]
  //      float f - frequencia em Hz - entre [0.1-50], default: 1.0 Hz
  //      int duty - duty cycle em % - entre [0-100], default: 50 %
  // Output parameter: none

  //parameter check
  if(pin<2||pin>13){ 
    Serial.println("Valor do pin errado");
    return;
  }
  if(f<0.1||f>50){ 
    Serial.println("Erro: Frequencia fora do intervalo [0.1-50]");
    Serial.println("Valor novo: f=1 Hz");
    f=1;
  }
  if(duty<0||duty>100){ 
    Serial.println("Erro: Duty cycle fora do intervalo [0-100]");
    Serial.println("Valor novo: duty=50%");
    duty=50;
  }

  //parametros de temporizacao 
  int T =1000/f;  //periodo em milisegundos
  int TH=T/100*duty, TL=T-TH;
  //Serial.println(duty);

  //parametros da maquina de estado
  const int APAGADO=0, ACESO=1;
  static int state=APAGADO;
  static unsigned long t0=millis();

  //estrutura máquina de estado

  switch(state){

    case APAGADO:
      if(millis()-t0>TL){
        digitalWrite(pin,HIGH);
        t0=millis();
        state=ACESO;
      }
    break;

    case ACESO:
      if(millis()-t0>TH){
        digitalWrite(pin,LOW);
        t0=millis();
        state=APAGADO;
      }
    break;   
  }
}   // end state machine 1

// state machine 2
void autoBlinkLed2(int pin,float f, int duty){
  // pisca um LED no pin com frequencia f
  // Input parameters: 
  //      int pin - numero pin do arduino - entre [2-13]
  //      float f - frequencia em Hz - entre [0.1-50], default: 1.0 Hz
  //      int duty - duty cycle em % - entre [0-100], default: 50 %
  // Output parameter: none

  //parameter check
  if(pin<2||pin>13){ 
    Serial.println("Valor do pin errado");
    return;
  }
  if(f<0.1||f>50){ 
    Serial.println("Erro: Frequencia fora do intervalo [0.1-50]");
    Serial.println("Valor novo: f=1 Hz");
    f=1;
  }
  if(duty<0||duty>100){ 
    Serial.println("Erro: Duty cycle fora do intervalo [0-100]");
    Serial.println("Valor novo: duty=50%");
    duty=50;
  }

  //parametros de temporizacao 
  int T =1000/f;  //periodo em milisegundos
  int TH=T/100*duty, TL=T-TH;
  //Serial.println(duty);

  //parametros da maquina de estado
  const int APAGADO=0, ACESO=1;
  static int state=APAGADO;
  static unsigned long t0=millis();

  //estrutura máquina de estado

  switch(state){

    case APAGADO:
      if(millis()>t0+TL){
        digitalWrite(pin,HIGH);
        t0=millis();
        state=ACESO;
      }
    break;

    case ACESO:
      if(millis()>t0+TH){
        digitalWrite(pin,LOW);
        t0=millis();
        state=APAGADO;
      }
    break;   
  }

}   // end state machine 2
```
